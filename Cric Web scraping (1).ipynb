```python
import requests
from bs4 import BeautifulSoup
```

```python
import numpy as np
import pandas as pd
import re
```

```python
request_header = {'Content-Type': 'text/html; charset=UTF-8','User-Agent': 'Chrome/101.0.0.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0','Accept-Encoding': 'gzip, deflate, br'}
```

```python
Player=[]
Matches=[]
Inns=[]
Not_outs=[]
Runs=[]
Ave=[]
B_Faced=[]
SR=[]
Fours=[]
Sixes=[]

for i in range(1,40):
    url = f"https://stats.espncricinfo.com/ci/engine/stats/index.html?class=3;page={i};spanmax2=31+Dec+2020;spanmin2=1+jan+2011;spanval2=span;template=results;type=batting"
    page = requests.get(url,headers = request_header)

    soup = BeautifulSoup(page.text)
    
    for i in soup.find_all('tr',class_="data1"):
        text = i.text

        regex=re.findall('[^\d\r-?:\r?\n|\r].+',text)
        if regex:
            Player.append(regex[0])
        else:
            Player.append(np.nan)

        regex=re.findall('\d+-\d+\s+(\d+)',text)
        if regex:
            Matches.append(regex[0])
        else:
            Matches.append(np.nan)

        regex=re.findall('\d+-\d+\s+\d+\s+(\d+)',text)
        if regex:
            Inns.append(regex[0])
        else:
            Inns.append(np.nan)

        regex=re.findall('\d+-\d+\s+\d+\s+\d+\s+(\d+)',text)
        if regex:
            Not_outs.append(regex[0])
        else:
            Not_outs.append(np.nan)

        regex=re.findall('\d+-\d+\s+\d+\s+\d+\s+\d+\s+(\d+)',text)
        if regex:
            Runs.append(regex[0])
        else:
            Runs.append(np.nan)

        regex=re.findall('\d+-\d+\s+\d+\s+\d+\s+\d+\s+\d+\s+\d.+\s+(\d.+)',text)
        if regex:
            Ave.append(regex[0])
        else:
            Ave.append(np.nan)

        regex=re.findall('\d+-\d+\s+\d+\s+\d+\s+\d+\s+\d+\s+\d.+\s+\d.+\s(\d+)',text)
        if regex:
            B_Faced.append(regex[0])
        else:
            B_Faced.append(np.nan)

        regex=re.findall('\d+-\d+\s+\d+\s+\d+\s+\d+\s+\d+\s+\d.+\s+\d.+\s\d.+\s+(\d.+)',text)
        if regex:
            SR.append(regex[0])
        else:
            SR.append(np.nan)

        regex=re.findall('(\d+)\s+\d+\s$',text)
        if regex:
            Fours.append(regex[0])
        else:
            Fours.append(np.nan)

        regex=re.findall('\d+\s+(\d+)\s$',text)
        if regex:
            Sixes.append(regex[0])
        else:
            Sixes.append(np.nan)
```

```python
print(len(Player))
print(len(Matches))
print(len(Inns))
print(len(Not_outs))
print(len(Runs))
print(len(Ave))
print(len(B_Faced))
print(len(SR))
print(len(Fours))
print(len(Sixes))
```

```python
Batters=pd.DataFrame({'Player':Player,'Matches':Matches,'Inns':Inns,'Not_outs':Not_outs,'Runs':Runs,'Ave':Ave,'B_Faced':B_Faced,'SR':SR,'Fours':Fours,'Sixes':Sixes})
```

```python
Batters.fillna(0,inplace=True)
```

```python
Batters=Batters.astype({'Matches':np.int16,'Inns':np.int16,'Not_outs':np.int16,'Runs':np.int16,'Ave':np.float16,'B_Faced':np.int16,'SR':np.float16,'Fours':np.int16,'Sixes':np.int16})
```

```python
Player=[]
Matches=[]
Inns=[]
Cts=[]
Sts=[]
D_per_Inns=[]
for i in range(1,26):
    url = f"https://stats.espncricinfo.com/ci/engine/stats/index.html?class=3;page={i};spanmax1=31+Dec+2020;spanmin1=1+jan+2011;spanval1=span;template=results;type=fielding"

    page = requests.get(url,headers = request_header)

    soup = BeautifulSoup(page.text)
    
    for i in soup.find_all('tr',class_="data1"):
        text = i.text

        regex=re.findall('([^\d\r-?:\r?\n|\r].+)\s+\d+-\d+',text)
        if regex:
            Player.append(regex[0])
        else:
            Player.append(np.nan)

        regex=re.findall('\d+-\d+\s+(\d+)',text)
        if regex:
            Matches.append(regex[0])
        else:
            Matches.append(np.nan)

        regex=re.findall('\d+-\d+\s+\d+\s+(\d+)',text)
        if regex:
            Inns.append(regex[0])
        else:
            Inns.append(np.nan)

        regex=re.findall('\d+-\d+\s+\d+\s+\d+\s+\d+\s+(\d+)',text)
        if regex:
            Cts.append(regex[0])
        else:
            Cts.append(np.nan)

        regex=re.findall('\d+-\d+\s+\d+\s+\d+\s+\d+\s+\d+\s+(\d+)',text)
        if regex:
            Sts.append(regex[0])
        else:
            Sts.append(np.nan)

        regex=re.findall('\d.\d+\s$',text)
        if regex:
            D_per_Inns.append(regex[0])
        else:
            D_per_Inns.append(np.nan)
```

```python
print(len(Player))
print(len(Matches))
print(len(Inns))
print(len(Cts))
print(len(Sts))
print(len(D_per_Inns))
```

```python
Wicket_keepers=pd.DataFrame({'Player':Player,'Matches':Matches,'Inns':Inns,'Cts':Cts,'Sts':Sts,'D_Ave_per_Inns':D_per_Inns})
```
